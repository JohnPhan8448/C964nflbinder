```python
import pandas as pd
import numpy as np
```

```python
def game_cleaning(df,stats):
    df['Tm'] = np.where(df.iloc[:, 5] == "@",df['Loser/tie'],df['Winner/tie'])
    df['away'] = np.where(df.iloc[:, 5] == "@",df['Winner/tie'],df['Loser/tie'])
    df['Tm_score'] = np.where(df["Tm"] == df["Winner/tie"],df["PtsW"],df["PtsL"])
    df['away_score'] = np.where(df["away"] == df["Winner/tie"],df["PtsW"],df["PtsL"])
    df['won'] = np.where((df['Tm_score'] > df['away_score']), True , False)
    col_names = ['Week', 'Date', 'away', 'away_score','Tm_score','Tm','won']
    temp = df.loc[:,col_names]
    temp['result'] = np.where((df['Tm_score'] > df['away_score']), 1, 0)
    cleaned = pd.merge(temp,stats,how="left",on= 'Tm')
    cleaned = pd.merge(cleaned,stats,left_on = 'away',right_on = 'Tm')
    cleaned['Date'] = pd.to_datetime(cleaned['Date'])
    cleaned = cleaned.sort_values("Date")
    return cleaned
```

```python
#Make sure you are using the correct games and stats before running!
df = pd.read_csv('2020games.csv')
stats = pd.read_csv('2020stats.csv')
cleaned = game_cleaning(df,stats)
cleaned
```

```python
cleaned.to_csv('week9cleaned.csv')
```
